# Has life expectancy increased over time in the six nations?

```python
plt.bar(df.Country,df['Life expectancy at birth (years)'])
plt.show()
```

```python
plt.plot(df.Country,df['Life expectancy at birth (years)'],marker='*')
plt.title('expectancy increased over time in the six nations')
plt.xlabel('Country')
plt.ylabel('Life Expectancy at birth (Years)')
plt.xticks(rotation=45)
plt.show()
```

```python
Chile = df[df.Country == "Chile"]
y1 = Chile["Life expectancy at birth (years)"]
China = df[df.Country == "China"]
y2 = China['Life expectancy at birth (years)']
```

```python
plt.plot(Chile.Year, y1)
plt.plot(China.Year, y2)
plt.show()
```

```python
sns.boxplot(data=df, x=df.Country, y=df['Life expectancy at birth (years)'] )
plt.xticks(rotation=45)
plt.show()
```

# Has GDP increased over time in the six nations?

```python
plt.plot(df.Country,df.GDP,marker='*')
plt.title('Country Vs GDP')
plt.xlabel('Country')
plt.ylabel('GDP')
plt.xticks(rotation=45)
plt.show()
```
